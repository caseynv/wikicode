```python
#print the qualifier and author names, given name, and family name.

def authorstring_author_info(list_item):
    site = pywikibot.Site("wikidata", "wikidata")
    repo = site.data_repository()
    for data_item in list_item:
        item = pywikibot.ItemPage(repo, data_item) 
        item_dict = item.get()
    
        for source in item.claims['P50']:
            QID = source.target.getID()
            item1 = pywikibot.ItemPage(repo, QID) 
            item_dict1 = item1.get()
            print(item_dict1['labels']['en'])
            print(item_dict1['descriptions']['en'])
            try:
                print(item_dict1['aliases']['en'])
            except:
                print('No description')
            try:
                item_new = item_dict1['claims']['P735']
                for item1_new in item_new:
                    itemm = item1_new.target.getID()
                    QID1 = pywikibot.ItemPage(repo, itemm) 
                    name = QID1.get()
                    print('Given name: ' + name['labels']['en'])
            except:
                print('No Given name')
            try:
                item_new1 = item_dict1['claims']['P734']
                for item1_new1 in item_new1:
                    itemm1 = item1_new1.target.getID()
                    QID2 = pywikibot.ItemPage(repo, itemm1) 
                    name1 = QID2.get()
                    print('Family name: ' + name1['labels']['en'])
            except:
                print('No Family name')
            tt = list(source.qualifiers.items())
            try:
                for key, value in tt:
                    try:
                        if key == 'P1545':
                            print('series ordinal: ' + value[0].getTarget())
                    except:
                        continue
                    try:
                        if key == 'P1932':
                            print('stated as: ' + value[0].getTarget())
                    except:
                        continue
                    try:
                        if key == 'P9688':
                            print('Last name - citation: ' + value[0].getTarget())
                    except:
                        continue
                    try:
                        if key == 'P9687':
                            print('Given name - citation: ' + value[0].getTarget())
                    except:
                        continue
            except:
                print('No qualifiers')
                print('\n')
                    
        try:
            for source in item.claims['P2093']:
                QID = source.target
                print(QID)
                stat = list(source.qualifiers.items())
                try:
                    for key, value in stat:
                        try:
                            if key == 'P1545':
                                print('series ordinal: ' + value[0].getTarget())
                        except:
                            continue
                        try:
                            if key == 'P9688':
                                print('Last name: ' + value[0].getTarget())
                        except:
                            continue
                        try:
                            if key == 'P9687':
                                print('Given name: ' + value[0].getTarget())
                        except:
                           continue
                except:
                    print('No qualifier')
        except:
            print('No author string')
                    
authorstring_author_info(['Q56603083', 'Q56603084', 'Q27450617', 'Q56942987', "Q27450018", "Q56603075"])
```
