```python
import pywikibot

#Get the item
site = pywikibot.Site("wikidata", "wikidata")
repo = site.data_repository()
item = pywikibot.ItemPage(repo, "Q27450018")

item_dict = item.get()
#Get the item dictionary
clm_dict = item_dict["claims"] # Get the claim dictionary
for clm in clm_dict:
    if 'P50' in clm:
        claim = pywikibot.Claim(repo, 'P50')
        for claim in item.claims['P50']: #Finds all statements (P131):
            p_value = claim.getTarget()
            p_item_dict = p_value.get()
            print(p_value.title())
            
            item1 = pywikibot.ItemPage(repo, p_value.title())
            item_dict1 = item1.get()
            #Get the item dictionary
            clm_dict1 = item_dict1["claims"] # Get the claim dictionary
            for clm1 in clm_dict1:
                print(clm1)
                print('\n*****************\n')
```

```python
#TODOS: Label the functions appropriately
#add text to a wikipedia/wikidata page

import pywikibot

site = pywikibot.Site('wikidata', 'wikidata')
repo = site.data_repository()
page = pywikibot.Page(site, 'User:Caseynv/Outreachy_1')

def edit(page):

    text = page.get()
    
    text = text + "\n  Hello"
    
    page.text = text
             
    try:
        page.save()
        return ('\n Saved! Thank you!')
    except:
        print("Oops! Try again")
        return 0
        
test = edit(page)
print(test)
```
